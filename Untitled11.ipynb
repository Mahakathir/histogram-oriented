```python
import cv2 as a
import matplotlib.pyplot as p
%matplotlib inline
#image processing library
from skimage.feature import hog as h
from skimage import data,exposure
```

```python
img=a.imread("child.jpeg")
p.imshow(img)
```

# HOG

```python
fd,hog_image=h(img,orientations=9,pixels_per_cell=(8,8),cells_per_block=(2,2),visualize=True,channel_axis=2)
```

# rescaling

```python
cale=exposure.rescale_intensity(hog_image,in_range=(0,10))
```

# plot

```python
figure,(a1,a2)=p.subplots(1,2,figsize=(12,6),sharex=True,sharey=True)
a1.axis('off')
a1.imshow(img)
a1.set_title('input image')

a2.axis('on')
a2.imshow(cale)
a2.set_title('histograme of oriented gradients')
p.show()
```

```python
h=a.HOGDescriptor()
h.setSVMDetector(a.HOGDescriptor_getDefaultPeopleDetector())
image=a.imread("people.jpeg")
image=imutils.resize(image,width=min(400,image.shape[1]))
(region,img)=h.detectMultiScale(image,winStride=(1,1),padding=(4,4),scale=1.05)
#drawing th region in the image
for(x,y,w,h) in region:
    a.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)
    
    
a.imshow("image",image)
a.waitKey(0)
a.destroyAllWindows()
    
```
